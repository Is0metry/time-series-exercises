Using the code from the lesson as a guide and the REST API from https://swapi.dev/ as we did in the lesson, create a dataframe named people that has all of the data for people

```python
def get_all_api(api:str,req:str)->pd.DataFrame:
    json = requests.get(api + req).json()
    results = json['results']
    while json['next'] is not None:
        json = requests.get(json['next']).json()
        results += json['results']
    return pd.DataFrame(results)
swapi = 'http://swapi.dev/api/'
get_all_swapi = lambda x: get_all_api(swapi,x)
people = get_all_swapi('people')
people.head()
```

Do the same thing, but for planets.


```python
planets = get_all_swapi('planets')
planets.head()
```

Extract the data for starships.


```python
starships = get_all_swapi('starships')
```

Save the data in your files to local csv files so that it will be faster to access in the future.


```python
people.to_csv('data/people.csv',index=False)
planets.to_csv('data/planets.csv',index=False)
starships.to_csv('data/starships.csv',index=False)
```

Combine the data from your three separate dataframes into one large dataframe.


```python
people_planets = people.merge(planets,left_on='homeworld',right_on='url')
people_planets.homeworld = people_planets.name_y
people_planets.name_x = people_planets.name_x.name = 'name'
people_planets.head()
def add_starships(ships:pd.Series):
    ship_names = []
    for ship in ships:
        ship_names.append(starships[starships.url == ship].name)
    return np.array(ship_names).flatten()
people_planets.starships = people_planets.starships.apply(add_starships)
star_wars = people_planets
star_wars.starships
```
